```
# import CSV file of linear data.
ls <- read.csv("https://raw.githubusercontent.com/17002235uhi/17002235uhi_DataAnalytics/main/lineardata3.csv", stringsAsFactors = TRUE)
```

```
head(ls)
```

```
summary(ls)
```

# Temperature

```
plot(ls$NUM_CRIMES, ls$temp, main="Does Temperature affect the Crime Rate in Chicago?")
```

```
tempresults <- lm(ls$temp ~ ls$NUM_CRIMES)
```

```
summary(tempresults)
```

```
plot(ls$NUM_CRIMES, ls$temp, main="Does Temperature affect Number of Crimes in Chicago?")
# draw line of best fit
abline(lm(ls$temp ~ ls$NUM_CRIMES))
```
