## Normalization of Number of Crimes

Due to the similar pattern over the years, I am going to do normalization scaling which rescaled the data into the range [0, 1]. This is to proportionalize the number of crimes data.

normalization: (𝑥 −𝑥𝑚𝑖𝑛)/(𝑥𝑚𝑎𝑥 −𝑥𝑚𝑖𝑛)

This code was sourced from: 
https://datasharkie.com/how-to-normalize-data-in-r/


```
summary(ttsc17d$NUM_CRIMES)
```

```
summary(ttsc18d$NUM_CRIMES)
```

```
summary(ttsc19d$NUM_CRIMES)
```

```
summary(ttsc20d$NUM_CRIMES)
```

```
normalize <- function(x, na.rm = TRUE) {
    return((x- min(x)) /(max(x)-min(x)))
}

ttsc17d$NUM_CRIMES<-normalize(ttsc17d$NUM_CRIMES)
ttsc18d$NUM_CRIMES<-normalize(ttsc18d$NUM_CRIMES)
ttsc19d$NUM_CRIMES<-normalize(ttsc19d$NUM_CRIMES)
ttsc20d$NUM_CRIMES<-normalize(ttsc20d$NUM_CRIMES)
```
